# Visualizing Company Sale data

### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# set this so the graphs open internally
%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://github.com/thieu1995/csv-files/blob/main/data/matplotlib/company_sales_data.csv).

```python
df = pd.read_csv('https://raw.githubusercontent.com/thieu1995/csv-files/main/data/matplotlib/company_sales_data.csv')
df
```

# Step 2b. Import data from columns

```python
month_list = df['month_number'].tolist()
face_cream = df['facecream'].tolist()
face_wash = df['facewash'].tolist()
tooth_paste = df['toothpaste'].tolist()
bathing_soap = df['bathingsoap'].tolist()
shampoo = df['shampoo'].tolist()
moisturizer = df['moisturizer'].tolist()
total_profit = df['total_profit'].tolist()
```

### Exercise 1: Read Total profit of all month_list and show it using a line plot

![img]()

```python
plt.figure(figsize=(8, 5))
plt.plot(month_list, total_profit, label='Profit data of last year',
         color='blue', markerfacecolor='black', linewidth=3)

plt.xlabel('Month Number')
plt.ylabel('Profit in dollar')
plt.title('Company Sales data of last year')
plt.xticks(month_list)
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.show()
```

### Exercise 2: Get total profit of all month_list and show line plot with the following Style properties

![img](img/02.png)

```python
plt.figure(figsize=(8, 5))
plt.plot(month_list, total_profit, label='Profit data of last year',
         color='red', marker='o', markerfacecolor='black',
         linestyle='--', linewidth=3)

plt.xlabel('Month Number')
plt.ylabel('Profit in dollar')
plt.legend(loc='lower right')
plt.title('Company Sales data of last year')
plt.xticks(month_list)
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.show()
```

### Exercise 3: Read all product sales data and show it  using a multiline plot

![img](img/03.png)

```python
plt.figure(figsize=(8, 5))

plt.plot(month_list, face_cream,   label='Face cream Sales Data', marker='o', linewidth=3)
plt.plot(month_list, face_wash,    label='Face Wash Sales Data',  marker='o', linewidth=3)
plt.plot(month_list, tooth_paste,  label='ToothPaste Sales Data', marker='o', linewidth=3)
plt.plot(month_list, bathing_soap, label='BathingSoap Sales Data', marker='o', linewidth=3)
plt.plot(month_list, shampoo,      label='Shampoo Sales Data',    marker='o', linewidth=3)
plt.plot(month_list, moisturizer,  label='Moisturizer Sales Data', marker='o', linewidth=3)

plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.title('Sales data')

plt.xticks(month_list)

plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])

plt.legend(loc='upper left')

plt.show()
```

### Exercise 4: Read toothpaste sales data of each month and show it using a scatter plot
![img](img/04.png)

```python
plt.figure(figsize=(8, 5))
plt.scatter(month_list, tooth_paste, label='Tooth paste Sales data')
plt.xlabel('Month Number')
plt.ylabel('Number of units Sold')
plt.title('Tooth paste Sales data')
plt.xticks(month_list)
plt.grid(True, linewidth=1, linestyle="--")
plt.legend(loc='upper left')
plt.show()
```

### Exercise 5: Read face cream and facewash product sales data and show it using the bar chart

![img](img/05.png)

```python
plt.figure(figsize=(7, 4.5))
plt.bar([x - 0.125 for x in month_list], face_cream, width=0.25, label='Face Cream sales data', align='center')
plt.bar([x + 0.125 for x in month_list], face_wash, width=0.25, label='Face Wash sales data', align='center')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.title('Facewash and facecream sales data')
plt.xticks(month_list)
plt.grid(True, linewidth=1, linestyle="--")
plt.legend(loc='upper left')
plt.show()
```

### Exercise 6: Read sales data of bathing soap of all month_list and show it using a bar chart. Save this plot to your hard disk

### Exercise 7: Read the total profit of each month and show it using the histogram to see the most common profit ranges

![img](img/07.png)

```python
profit_range = [150000, 175000, 200000, 225000, 250000, 300000, 350000]
plt.figure(figsize=(7, 4.5))
plt.hist(total_profit, bins=profit_range, label='Profit data')
plt.xlabel('profit range in dollar')
plt.ylabel('Actual Profit in dollar')
plt.title('Profit data')
plt.xticks(profit_range)
plt.legend(loc='upper left')
plt.show()
```

### Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart

![img](img/08.png)

```python
labels = ['FaceCream', 'FaseWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer']
sales_data = [df['facecream'].sum(), df['facewash'].sum(), df['toothpaste'].sum(),
              df['bathingsoap'].sum(), df['shampoo'].sum(), df['moisturizer'].sum()]
plt.figure(figsize=(7, 7))
plt.pie(sales_data, labels=labels, autopct='%1.1f%%')
plt.title('Sales data')
plt.legend(loc='lower right')
plt.show()
```

### Exercise 9: Read Bathing soap facewash of all month_list and display it using the Subplot

![img](img/09.png)

```python
f, axarr = plt.subplots(2, sharex=True, figsize=(7, 5))
axarr[0].plot(month_list, bathing_soap, label='Bathingsoap Sales Data', color='k', marker='o', linewidth=3)
axarr[0].set_title('Sales data of  a Bathingsoap')
axarr[1].plot(month_list, face_wash, label='Face Wash Sales Data', color='r', marker='o', linewidth=3)
axarr[1].set_title('Sales data of  a facewash')
plt.xticks(month_list)
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.show()
```

### Exercise 10: Read all product sales data and show it using the stack plot

![img](img/10.png)

```python
plt.figure(figsize=(8, 5))

plt.stackplot(month_list,
              face_cream, face_wash, tooth_paste,
              bathing_soap, shampoo, moisturizer,
              labels=['face Cream', 'Face wash', 'Tooth paste', 'Bathing soap', 'Shampoo', 'Moisturizer'],
              colors=['m', 'c', 'r', 'k', 'g', 'y'])

plt.title('Alll product sales data using stack plot')
plt.xlabel('Month Number')
plt.ylabel('Sales unints in Number')

plt.legend(loc='upper left')

plt.xticks(month_list)

plt.ylim(0, 31000)

plt.show()
```
