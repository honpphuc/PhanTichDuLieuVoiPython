### Step 6. Is there more male or female names in the dataset?

```python
gender_counts = baby_names['Gender'].value_counts()
if(gender_counts['F'] > gender_counts['M']):
  print("There are more female name")
elif(gender_counts['F'] == gender_counts['M']):
  print("There are more male name")
else:
  print("Male and Female names are the same")
```

### Step 7. Group the dataset by name and assign to names

```python
names = baby_names['Name']
names
```

### Step 8. How many different names exist in the dataset?

```python
counts = baby_names['Name'].nunique()
print(f"Co {counts} ten khac nhau")
```

### Step 9. What is the name with most occurrences?

```python
print(baby_names.groupby('Name').Count.sum().idxmax())
```

### Step 10. How many different names have the least occurrences?

```python
print(f"The name which have the least occurrences is {baby_names.groupby('Name').Count.sum().idxmin()} - {baby_names.groupby('Name').Count.sum().min()} times")
```

### Step 11. What is the median name occurrence?




```python
df3 = baby_names.groupby('Name').Count.sum()
median = df3.median()
print(f"The median name occurrence is {median}")
```

### Step 12. What is the standard deviation of names?

```python
print(f"The standard deviation of name is: {df3.std()}")
```
