### Step 3. Assign it to a variable called euro12.

```python
euro12 = pd.read_csv('https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/Euro_2012_stats_TEAM.csv',index_col=0,low_memory=False)
euro12
```

### Step 4. Select only the Goal column.

```python
print(f"There are {euro12.shape[0]} teams which took part in Euro2012")
```

```python
print(f"There are: {euro12.shape[1]} columns")
```

### Step 7. View only the columns Team, Yellow Cards and Red Cards and assign them to a dataframe called discipline

```python
discipline = euro12[['Yellow Cards', 'Red Cards']]
discipline
```

### Step 8. Sort the teams by Red Cards, then to Yellow Cards

```python
#Sort the teams by Red Cards
df_redcards = euro12.sort_values('Red Cards')
df_redcards
```

```python
#Sort the teams by Yellow Cards
df_yellowcards = euro12.sort_values('Yellow Cards')
df_yellowcards
```

### Step 9. Calculate the mean Yellow Cards given per Team



```python
print(f"The mean Yellow Cards given per Team is: {discipline['Yellow Cards'].mean()}")
```
